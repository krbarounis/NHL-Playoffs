```python
def get_game_stats(games):
    """get_game_stats(games):
    Build a list of game "rows";
    that is, a list of dictionaries
    containing game data that we care about:
        gameID
        home team
        home team id
        away team
        away team id
        and all game stats per team
    Parameters:
        games:  list of full game dictionaries returned
        from API
    Returns:
        game_rows (list of dicts)
    """
    # empty list to hold our dicts
    game_rows = []
    for gameId in games:
        BASE_URL = 'https://statsapi.web.nhl.com/api/v1/game/{}/boxscore'
        game_url = BASE_URL.format(gameId)
        response = requests.get(game_url)
        game_data = response.json()
      
        home_team_id = game_data['teams']['home']['team']['id']
        home_team = game_data['teams']['home']['team']['name']
        home_goals = game_data['teams']['home']['teamStats']['teamSkaterStats']['goals']
        home_pim = game_data['teams']['home']['teamStats']['teamSkaterStats']['pim']
        home_shots = game_data['teams']['home']['teamStats']['teamSkaterStats']['shots']
        home_pp_pct = game_data['teams']['home']['teamStats']['teamSkaterStats']['powerPlayPercentage']
        home_pp_goals = game_data['teams']['home']['teamStats']['teamSkaterStats']['powerPlayGoals'] 
        home_pp_opp = game_data['teams']['home']['teamStats']['teamSkaterStats']['powerPlayOpportunities']         
        home_face_win_pct = game_data['teams']['home']['teamStats']['teamSkaterStats']['faceOffWinPercentage']         
        home_blocked = game_data['teams']['home']['teamStats']['teamSkaterStats']['blocked']         
        home_takeaways = game_data['teams']['home']['teamStats']['teamSkaterStats']['takeaways']                 
        home_giveaways = game_data['teams']['home']['teamStats']['teamSkaterStats']['giveaways']                 
        home_hits = game_data['teams']['home']['teamStats']['teamSkaterStats']['hits']  
        
        away_team_id = game_data['teams']['away']['team']['id']
        away_team = game_data['teams']['away']['team']['name']
        away_goals = game_data['teams']['away']['teamStats']['teamSkaterStats']['goals']
        away_pim = game_data['teams']['away']['teamStats']['teamSkaterStats']['pim']
        away_shots = game_data['teams']['away']['teamStats']['teamSkaterStats']['shots']
        away_pp_pct = game_data['teams']['away']['teamStats']['teamSkaterStats']['powerPlayPercentage']
        away_pp_goals = game_data['teams']['away']['teamStats']['teamSkaterStats']['powerPlayGoals'] 
        away_pp_opp = game_data['teams']['away']['teamStats']['teamSkaterStats']['powerPlayOpportunities']         
        away_face_win_pct = game_data['teams']['away']['teamStats']['teamSkaterStats']['faceOffWinPercentage']         
        away_blocked = game_data['teams']['away']['teamStats']['teamSkaterStats']['blocked']         
        away_takeaways = game_data['teams']['away']['teamStats']['teamSkaterStats']['takeaways']                 
        away_giveaways = game_data['teams']['away']['teamStats']['teamSkaterStats']['giveaways']                 
        away_hits = game_data['teams']['away']['teamStats']['teamSkaterStats']['hits']
        
        home_win = 1 if home_goals > away_goals else 0
        away_win = 1 if away_goals > home_goals else 0        
        
        game_rows.append(dict(gameId=gameId,
                            home_team_id=home_team_id,
                            home_team=home_team,
                            home_goals=home_goals,
                            home_pim=home_pim,
                            home_shots=home_shots,
                            home_pp_pct=home_pp_pct,
                            home_pp_goals=home_pp_goals,
                            home_pp_opp=home_pp_opp,
                            home_face_win_pct=home_face_win_pct,
                            home_blocked=home_blocked,
                            home_takeaways=home_takeaways,
                            home_giveaways=home_giveaways,
                            home_hits=home_hits,
                            home_win=home_win,
                            away_team_id=away_team_id,
                            away_team=away_team,
                            away_goals=away_goals,
                            away_pim=away_pim,
                            away_shots=away_shots,
                            away_pp_pct=away_pp_pct,
                            away_pp_goals=away_pp_goals,
                            away_pp_opp=away_pp_opp,
                            away_face_win_pct=away_face_win_pct,
                            away_blocked=away_blocked,
                            away_takeaways=away_takeaways,
                            away_giveaways=away_giveaways,
                            away_hits=away_hits,
                            away_win=away_win))
        
    return game_rows
```

```python
###
### NEW DIRECTION:
### 1) Get all playoff gamese for a given season;
### 2) For each playoff game:
###     - Get all team IDs
###     - Get unique set of team ids (per season)
### 3) For each team ID:
###    - Get all regular season stats
###
```

```python
def get_playoff_games(season):
    BASE_URL = 'https://records.nhl.com/site/api/playoff-series?cayenneExp=seasonId={}'
    playoff_url = BASE_URL.format(season)
    
    response = requests.get(playoff_url)
    playoff_data = response.json()
    
    return [game['gameId'] for game in playoff_data['data']]
    
```

```python
def get_team_ids(game_ids):
    # empty list to hold our dicts
    team_ids = []
    for gameId in game_ids:
        BASE_URL = 'https://statsapi.web.nhl.com/api/v1/game/{}/boxscore'
        game_url = BASE_URL.format(gameId)
        response = requests.get(game_url)
        game_data = response.json()
        
        # only need "home" games because each team plays at home
        # at least half of a playoff series
        team_ids.append(game_data['teams']['home']['team']['id'])
        
    # make unique
    return set(team_ids)
```

```python
# ALL PLAYOFF SEASONS WE ARE INTERESTED IN:
seasons = ['20182019',
'20172018',
'20162017',
'20152016',
'20142015',
'20132014',
'20122013',
'20112012',
'20102011',
'20092010',
'20082009',
'20072008']
```

```python
#
# for each season:
# call get_playoff_games to get all game ids:
# for each game id:
# get all UNIQUE team ids from all of that season;s
# playoff games
#
# for each unique set of team ids:
# create dict of team ids : season

list_of_playoff_teams = []
for season in seasons:
    playoff_games = get_playoff_games(season)
    team_ids = get_team_ids(playoff_games)
    season_teams = dict.fromkeys(team_ids, season)
    list_of_playoff_teams.append(season_teams)
    
```
